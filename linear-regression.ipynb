```python
v=pd.read_csv("/kaggle/input/single-feature-linear-regression-dataset/Data.csv")
```

```python
v.head()
```

```python
#y=mx+b
r=v
```

```python
r['xy']=r['Feature']*r['Target']
r['xsquare']= r['Feature']*r['Feature']
sumx2=sum(r['xsquare'])
sumxy=sum(r['xy'])
sumx=sum(r['Feature'])
sumx
sumy=sum(r['Target'])
sumy
f=(sumy*sumx2)-(sumx*sumxy)
y2=100*(sumx2)-(sumx*sumx)
alpha=f/y2
alpha
beta=(100*(sumxy)-sumx*sumy)/(100*(sumx2)-(sumx*sumx))
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
```

```python
l
```

```python
x=r[['Feature']]
y=r['Target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
len(x_train),len(y_train)
```

```python
l.fit(x_train,y_train)
```

```python
e=l.predict(x_test)
len(e)
```

```python
import pandas as pd
ww=[]
for i in x_test['Feature']:
    ww.append((i*beta) + alpha)
print(ww)
```

```python
ven=l.predict(x_test)
```
