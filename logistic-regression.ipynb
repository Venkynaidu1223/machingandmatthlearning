```python
x=v[['Model', 'Year', 'Region', 'Color', 'Fuel_Type', 'Transmission','Engine_Size_L', 'Mileage_KM', 'Price_USD', 'Sales_Volume']]
y=v[['Sales_Classification']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg
```

```python
lg.fit(x_train,y_train)
```

```python
ss=lg.predict(x_test)
ss
```

```python
lg.score(x_test,y_test)
```

```python
ap=pd.DataFrame({'predict':ss})
```

```python
ap['real'] = y_test['Sales_Classification'].reset_index(drop=True)
```

```python
ap
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(ap['real'], ap['predict'])

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
